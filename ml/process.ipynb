```python
from segment_anything import SamAutomaticMaskGenerator, sam_model_registry

sam = sam_model_registry["vit_h"](checkpoint="sam_vit_h_4b8939.pth")
mask_generator = SamAutomaticMaskGenerator(sam)
```

```python
# Function that inputs the output and plots image and mask
from matplotlib import pyplot as plt
import numpy as np

# https://www.freecodecamp.org/news/use-segment-anything-model-to-create-masks/


def show_output(result_dict, axes=None):
    if axes:
        ax = axes
    else:
        ax = plt.gca()
        ax.set_autoscale_on(False)
    sorted_result = sorted(result_dict, key=(
        lambda x: x['area']),      reverse=True)
    # Plot for each segment area
    for val in sorted_result:
        mask = val['segmentation']
        img = np.ones((mask.shape[0], mask.shape[1], 3))
        color_mask = np.random.random((1, 3)).tolist()[0]
        for i in range(3):
            img[:, :, i] = color_mask[i]
            ax.imshow(np.dstack((img, mask*0.5)))
```

```python
import datetime
import os
import cv2
import numpy as np

subdirectory = "./images"

# loop all files in subdirectory
i = 0
for filename in os.listdir(subdirectory):
    i += 1
    if i > 2:
        break
    image = cv2.imread(os.path.join(subdirectory, filename))
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    t = datetime.datetime.now()
    output_mask = mask_generator.generate(image_rgb)
    print("", datetime.datetime.now() - t)

    _,axes = plt.subplots(1,2, figsize=(16,16))
    axes[0].imshow(image_rgb)
    show_output(output_mask, axes[1])
```
